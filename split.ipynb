```python
import pandas as pd
from sklearn.model_selection import train_test_split
import pyarrow.parquet as pq
import pyarrow as pa

# Assuming df is your original DataFrame
filepath = "./data/catB_train.parquet"
df = pd.read_parquet(filepath)

# Splitting the DataFrame into 70% train and 30% test
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)

# Exporting train and test DataFrames to parquet files
train_filepath = "./data/train_data.parquet"
test_filepath = "./data/test_data.parquet"

# Writing DataFrames to parquet files
train_table = pa.Table.from_pandas(train_df)
test_table = pa.Table.from_pandas(test_df)

pq.write_table(train_table, train_filepath)
pq.write_table(test_table, test_filepath)
```
